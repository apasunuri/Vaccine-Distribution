```python
correlation_df = df.drop(columns=["FIPS", "Countyname", "ST_Name", "ST_Abbr"])
```

```python
for i in correlation_df.columns.values:
    print(correlation_df[i].dtype)
```

```python
usable = correlation_df.corr()
```

```python
sns.heatmap(usable)
```

```python
strong = usable[abs(usable["Deaths"]) >= 0.9]
```

```python
strong
```

```python
square_strong = strong[strong.index]
```

```python
sns.heatmap(square_strong)
```

```python
dfMax = correlation_df.max()
scaled = correlation_df.divide(dfMax, axis=1)
```

```python
scaled
```

```python
s_corr = scaled.corr()
```

```python
s_corr
```

```python
scaled_strong = s_corr[abs(s_corr["Deaths"]) >= 0.9]
```
