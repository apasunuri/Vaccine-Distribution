```python
poverty = pd.read_excel("../raw-data/County-Poverty.xls")
poverty.head()
```

```python
cols = poverty.iloc[2].values
poverty.columns = cols
```

```python
poverty = poverty.iloc[3:]
poverty.head()
```

```python
poverty["FIPS"] = str(poverty["State FIPS Code"]).strip() + str(poverty["County FIPS Code"]).strip()
```

```python
poverty["FIPS"] = poverty["State FIPS Code"] + poverty["County FIPS Code"]
```

```python
poverty["FIPS"] = poverty["FIPS"].apply(lambda x: x[1:] if x[0] == "0" else x)
```

```python
poverty.head()
```

```python
poverty = poverty[["FIPS", "Poverty Estimate, All Ages"]]
```

```python
poverty["FIPS"] = poverty["FIPS"].apply(lambda x: int(x))
```

```python
test = county_covid.set_index("FIPS").join(poverty.set_index("FIPS"))
```

```python
test[test["Countyname"] == "Alachua"]
```

```python
hosp_beds["FIPS"] = hosp_beds["FIPS"].apply(lambda x: 0 if (x != x) else int(x))
```

```python
hosp = hosp_beds.groupby("FIPS").agg({"NUM_LICENSED_BEDS" : np.sum, "NUM_STAFFED_BEDS" : np.sum, "NUM_ICU_BEDS" : np.sum,
                              "ADULT_ICU_BEDS" : np.sum, "PEDI_ICU_BEDS" : np.sum, "BED_UTILIZATION" : np.average,
                              "AVG_VENTILATOR_USAGE" : np.average, "Potential_Increase_In_Bed_Capac" : np.sum})
```
