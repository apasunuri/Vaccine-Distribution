```python
df.drop(["OBJECTID", "ST_Abbr", "ST_ID", "PCTPOVALL_", "State_Fata", "DateChecke", "EM_type", "EM_date",
        "EM_notes", "url", "Thumbnail", "Age_85", "Age_80_84", "Age_75_79", "Age_70_74", "Age_65_69",
        "Beds_Licen", "Beds_Staff", "POP_ESTIMA", "POVALL_201", "Median_Hou", "State_Conf", "State_Deat", "State_Reco",
        "AgedPop", "TotalPop", "TwoMorPop", "Wh_Alone", "Bk_Alone", "AI_Alone", "As_Alone", "NH_Alone", "SO_Alone",
        "Two_More", "Not_Hisp", "Age_Less15", "Age_15_24", "Age_25_34", "Age_Over75", "Agetotal", "NonHisp", 
        "Age_35_64", "Age_65_74", "Day_1", "Day_2", "Day_3", "Day_4", "Day_5", "Day_6", "Day_7", "Day_8", "Day_9",
        "Day_10", "Day_11", "Day_12", "Day_13", "Day_14", "Inpat_Occ", "ICU_Occ", "Inpat_Cap", "Inpat_COV", "Inp_nonCOV",
        "ICU_COV", "ADULT_ICU_BEDS", "PEDI_ICU_BEDS"], axis = 1, inplace = True)
```

```python
len(df.columns.values)
```

```python
df.columns.values
```

```python
df["Population_Density"] = df["Population"] / df["Shape__Area"]
df["ICU_Beds"] = df["NUM_ICU_BEDS"] / df["Population"]
df["ICU_Capacity"] = df["ICU_Cap"] / df["Population"]
df["Unemployment"] = df["Unemployed"] / df["Population"]
df["Non_Hisp"] = df["NonHispWhP"] / df["Population"]
df["Black"] = df["BlackPop"] / df["Population"]
df["AmInd"] = df["AmIndop"] / df["Population"]
df["Asian"] = df["AsianPop"] / df["Population"]
df["PacIsl"] = df["PacIslPop"] / df["Population"]
df["OtherPop"] = df["OtherPop"] / df["Population"]
df["Hisp"] = df["HispPop"] / df["Population"]
#df["Poverty Estimate, All Ages"].apply(lambda x: 0 if x == '.' else x)
#df["Poverty Estimate, All Ages"] = df["Poverty Estimate, All Ages"].astype(float)
df["Poverty Estimate, All Ages"] = df["Poverty Estimate, All Ages"].replace({".", 0})
df["Poverty Estimate, All Ages"] = df["Poverty Estimate, All Ages"].astype(int)
df["Poverty"] = df["Poverty Estimate, All Ages"] / df["Population"]
df["NewDeaths"] = df["NewDeaths"] / df["Population"]
df["NewCases"] = df["NewCases"] / df["Population"]
df["Ventilator"] = df["Ventilator"] / df["Population"]
df["RecoveredByP"] = df["Recovered"] / df["Population"]
df["ActiveByP"] = df["Active"] / df["Population"]
df.drop(["NUM_ICU_BEDS", "ICU_Cap", "Unemployed", "NonHispWhP", "BlackPop", "AmIndop", "AsianPop", "PacIslPop",
        "OtherPop", "HispPop", "Poverty Estimate, All Ages"], inplace = True, axis = 1)
```

```python
df["NewDeaths"] = df["NewDeaths"] / df["Population"]
df["NewCases"] = df["NewCases"] / df["Population"]
df["Ventilator"] = df["Ventilator"] / df["Population"]
df["RecoveredByP"] = df["Recovered"] / df["Population"]
df["ActiveByP"] = df["Active"] / df["Population"]
```
